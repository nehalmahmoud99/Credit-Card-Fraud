```python
data = data.drop(columns = ['Time'], axis = 1)
data
```

let's train , test data  
# first: spilt the data  
# then : train model 

```python
X = data.drop(columns = 'Class', axis = 1)
y= data['Class']
 
```

```python
X
```

```python
y
```

```python


from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1 )

```

```python


print("Shape of X_train ", X_train.shape)
print("Shape of X_test ", X_test.shape)
print("Shape of y_train ", y_train.shape)
print("Shape of y_test ", y_test.shape)
```

```python
# Standardizing  
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

 MODEL THE DATA 

```python

from sklearn.metrics import accuracy_score, precision_score, confusion_matrix, recall_score, f1_score

from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier 

from sklearn.tree import DecisionTreeClassifier 

import warnings

# Ignore warnings
warnings.filterwarnings("ignore")


```

```python

# Logistic Regression
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 123)
classifier.fit(X_train, y_train)

# Predicting the accuracy of the model

from sklearn.metrics import confusion_matrix, accuracy_score
y_pred = classifier.predict(X_test)
cm = confusion_matrix(y_test, y_pred)

accuracy_score(y_test, y_pred
              ) * 100


```

```python
# Implementing Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
D_S = DecisionTreeClassifier()
D_S =D_S.fit(X_train, y_train)

# Predicting the accuracy of the model

y_pred2 = D_S.predict(X_test)
accuracy_score(y_test, y_pred2) * 100 




```

```python
# Implementing Random Forest
from sklearn.ensemble import RandomForestClassifier
Random = RandomForestClassifier(max_depth=2, random_state=0, n_estimators=100)
Random.fit(X_train, y_train)


# Getting the accuracy of the model

y_pred3 = Random.predict(X_test)
accuracy_score(y_test, y_pred3) * 100
```

```python
# Applying Gradient Boosting Classifier 
gradient = GradientBoostingClassifier(n_estimators = 100, random_state = 0)
gradient.fit(X_train, y_train)  

# Getting the accuracy of the model
y_pred4= gradient.predict(X_test) 
accuracy_score(y_test, y_pred4) *100
```

As we can see with our DecisionTree   Model we are getting a better result even for the recall which is the most tricky part.
